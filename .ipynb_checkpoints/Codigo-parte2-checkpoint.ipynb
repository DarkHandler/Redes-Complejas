### IMPORTS

```python
import networkx as nx
import pandas as pd
import random
import collections
import matplotlib.pyplot as plt
import numpy as np
```

### En esta simulación se usa solamente los datos del total de personas infectadas hasta la fecha, para hacer la simulación en adelante 

```python
regiones = ['Arica y Parinacota','Tarapacá','Antofagasta','Atacama','Coquimbo','Valparaíso','Metropolitana','O’Higgins','Maule','Ñuble','Biobío','Araucanía','Los Ríos','Los Lagos','Aysén','Magallanes']
archivo_c = pd.read_csv('Datos-COVID19/output/producto26/CasosNuevosConSintomas.csv') #colocar la ruta donde a usted le acomode
archivo_c = archivo_c.iloc[:, -1]
archivo_c = archivo_c.loc[[regiones.index("Coquimbo"),regiones.index("Valparaíso"),regiones.index("Metropolitana")]]

archivo_a = pd.read_csv('Datos-COVID19/output/producto27/CasosNuevosSinSintomas.csv') #colocar la ruta donde a usted le acomode
archivo_a = archivo_a.iloc[:, -1]
archivo_a = archivo_a.loc[[regiones.index("Coquimbo"),regiones.index("Valparaíso"),regiones.index("Metropolitana")]]
```

```python
from random import randrange
acro = ["C","V","S"]
move = [ 0 , 1 , 2 ]
def crearContagios(G,cant,estado):
    for i in range(1, cant+1):
        random = randrange(1, G.number_of_nodes())
        G.nodes[random]["Estado"] = estado
        G.nodes[random]["tinfeccion"] = 14
        colorear(G,random)
        if(estado == 1 and len(list(G.adj[random]))>1):
            reducirCon(G,random)
def definir(G,r):
    for i in G.nodes():
        G.nodes[i]["Estado"]= 0
        G.nodes[i]["Region"]= r
        G.nodes[i]["tinfeccion"]= 0
        G.nodes[i]["pcuarentena"] = []
        colorear(G,i)
def establecerGrafos(CH):
    for i in range(3):
        definir(CH[i],acro[i])
        crearContagios(CH[i],archivo_c.iloc[i],1)#1 con sintomas
        crearContagios(CH[i],archivo_a.iloc[i],2)#2 sin sintomas
    simulacionDias(5,CH)
    
def colorear(g,n):
    if g.nodes[n]["Estado"] == 0:
        g.nodes[n]["Color_estado"] = "green"
    elif g.nodes[n]["Estado"] == 1:
        g.nodes[n]["Color_estado"] = "red"
    else:
        g.nodes[n]["Color_estado"] = "yellow"
    
    if g.nodes[n]["Region"] == "C":
        g.nodes[n]["Color_region"] = "blue"
    elif g.nodes[n]["Region"] == "S":
        g.nodes[n]["Color_region"] = "grey"
    else:
        g.nodes[n]["Color_region"] = "pink"
```

```python
def trasladoNodo(origen,destino,nodo): #translada en nodo a un destino y copia datos que tenia en el origen
    new = origen.nodes[nodo]["Region"]+str(nodo)
    destino.add_node(new)
    destino.nodes[new]["Estado"] = origen.nodes[nodo]["Estado"]
    destino.nodes[new]["Region"] = origen.nodes[nodo]["Region"]
    destino.nodes[new]["tinfeccion"] = origen.nodes[nodo]["tinfeccion"]
    destino.nodes[new]["pcuarentena"] = origen.nodes[nodo]["pcuarentena"]
    destino.nodes[new]["Color_estado"] = origen.nodes[nodo]["Color_estado"]
    destino.nodes[new]["Color_region"] = origen.nodes[nodo]["Color_region"]
    destino.nodes[new]["oldList"] = list(origen.adj[nodo])
    destino.nodes[new]["oldubi"] = nodo
    vecino = random.choice(list(destino.nodes()))
    crearEdge(destino,new,vecino)
    
def crearEdge(destino,nodo,vecino): #crear las aristas vecinas que tendra cuando viaje al otro grafo (region)
    if(isinstance(vecino,int)):
        destino.add_edge(nodo,vecino)
    else:
        vecino = random.choice(list(destino.nodes()))
        crearEdge(destino,nodo,vecino)

def reducirCon(G,nodo): #reduce los vecinos que tiene un ConSintomas cuando esta en cuarentena
    old_vecinos = list(G.adj[nodo])
    new_vecinos = []
    for i in old_vecinos:
        if(isinstance(i,int)):
            new_vecinos.append(i)
    if(len(new_vecinos) > 2):
        cant = int(len(new_vecinos)*0.7)
        for i in range(cant):
            rand = random.choice(new_vecinos)
            listvecinos = list(G.adj[rand])
            listvecinos.remove(nodo)
            for i in listvecinos:
                if(isinstance(i,str)):
                    listvecinos.remove(i)
            if(len(listvecinos) > 2):
                G.nodes[nodo]["pcuarentena"].append(rand)
                G.remove_edge(nodo,rand)
                new_vecinos.remove(rand)
    
def restablecerCon(G,nodo): #restablece los vecinos que perdio por haber estado en cuarentena (el ConSintomas)
    old = G.nodes[nodo]["pcuarentena"]
    for i in old:
        if(i != "null"):
            if(i in G):
                G.add_edge(nodo,i)
    G.nodes[nodo]["pcuarentena"] = []
def eliminarNodos(G,n):
    if(n):
        for i in n:
            G.remove_node(i)
def retornoNodo(G,n): #agrega todos los atributos que tenia el nodo antes de haber viajado a otra region
    retorno = G.nodes[n]["Region"]
    ubi = acro.index(retorno)
    old_n = G.nodes[n]["oldubi"] #numero del nodo antiguo
    CH[ubi].add_node(old_n)
    CH[ubi].nodes[old_n]["Estado"] = G.nodes[n]["Estado"]
    CH[ubi].nodes[old_n]["Region"] = G.nodes[n]["Region"]
    CH[ubi].nodes[old_n]["tinfeccion"] = G.nodes[n]["tinfeccion"]
    CH[ubi].nodes[old_n]["pcuarentena"] = G.nodes[n]["pcuarentena"]
    CH[ubi].nodes[old_n]["Color_estado"] = G.nodes[n]["Color_estado"]
    CH[ubi].nodes[old_n]["Color_region"] = G.nodes[n]["Color_region"]
    
    for v in G.nodes[n]["oldList"]:
        if(v in CH[ubi]):
            CH[ubi].add_edge(old_n,v)
        else:
            CH[ubi].nodes[old_n]["pcuarentena"].append(v)
```

```python
acro = ["C","V","S"]
move = [ 0 , 1 , 2 ]
def simulacionDias(dias,CH):
    analizar = Analizador()
    analizar.setPositionGraph(CH) #esto es para posicionar a los nodos en las imagenes que se generan
    
    for d in range(dias):
        nuevosInfectados = 0
        recuperados = 0
        for g in range(3):
            nuevosInfectados_region = 0
            for n in CH[g].nodes():
                if(CH[g].nodes[n]["Estado"] == 1 or CH[g].nodes[n]["Estado"] == 2):#revisar si tiene infeccion
                    CH[g].nodes[n]["tinfeccion"]-=1#reducir infeccion
                    if(CH[g].nodes[n]["tinfeccion"]==0):#si llega a cero se sana
                        if(CH[g].nodes[n]["Estado"] == 1):#si estuvo en cuarentena
                            restablecerCon(CH[g],n)
                        CH[g].nodes[n]["Estado"] = 0
                        recuperados+=1
                else:
                    if(CH[g].nodes[n]["pcuarentena"]):
                        restablecerCon(CH[g],n)
                    vecinos = list(CH[g].adj[n])
                    cant = 0
                    if(vecinos):#si tiene vecinos
                        for v in vecinos:#recorre los vecinos
                            if(CH[g].nodes[v]["Estado"] == 1 or CH[g].nodes[v]["Estado"] == 2):
                                cant+=1
                        prob = cant/len(vecinos)
                        rand = random.random()
                        if(rand<prob):
                            if(random.random()<0.45):#probabilidad de infectar sin sintomas 0.45
                                CH[g].nodes[n]["Estado"] = 2
                                CH[g].nodes[n]["tinfeccion"]=14
                            else:
                                CH[g].nodes[n]["Estado"] = 1
                                CH[g].nodes[n]["tinfeccion"]=14
                                reducirCon(CH[g],n)
                            colorear(CH[g],n)
                            nuevosInfectados+=1
                            nuevosInfectados_region+=1
            analizar.agregar(nuevosInfectados_region)
            print()
            print("Region",acro[g],"nuevos infectados:",nuevosInfectados)
            print("Region",acro[g],"recuperados:",recuperados)
            print()
        for g in range(3):
            nodosold = []#nodos movidos para eleminar del orgien
            cont_move = 0
            cont_return = 0
            for n in CH[g].nodes():
                valid = True
                if(CH[g].nodes[n]["Estado"] == 2):#solo se mueve gente sin sintomas
                    for v in list(CH[g].adj[n]):
                        listvecinos = list(CH[g].adj[v])
                        listvecinos.remove(n)
                        for i in listvecinos:
                            if(isinstance(i,str)):
                                listvecinos.remove(i)
                        if(len(listvecinos)<=2):
                            valid = False
                    if(random.random() < 0.5 and valid):#probabilidad de moverte
                        if(CH[g].nodes[n]["Region"] != acro[g] and valid):#si ya se movio regresa
                            retornoNodo(CH[g],n)
                            nodosold.append(n)
                            cont_return = 0
                        else:#si no se mueve a otro lado
                            newlist = move.copy()#copia lista de ubicaciones de region
                            newlist.remove(g)#remueve region actual
                            region_new = random.choice(newlist)#escoje region aleatoria
                            trasladoNodo(CH[g],CH[region_new],n)
                            nodosold.append(n)
                            cont_move+=1
            eliminarNodos(CH[g],nodosold)
            print("Movimiento: ",cont_move,"Retornos: ",cont_return)
        dia=str(d+1) 
        analizar.analizarEstados(CH)
        print("Fin dia "+dia)
        #se deben dibujar los 3 grafos cada dia y ser guardados
        #hago un reset de las posiciones de cada grafo
        analizar.resetPositionGraph(CH)
        analizar.drawGraph(CH,dia)
        
        print("========================================================================================")
    analizar.printGraficos()
    


class Analizador: #analizar grafo cuando se le es pasado

    def __init__(self):
        self.historico = []
        self.total = []
        self.hist_nuevo = []
        self.hist_diametro=[]
        self.posGraph=[] #guardara posicion de nodos en grafo
    
    def analizarEstados(self,CH):
        for G in CH:
            #print(nx.is_connected(G))
            cant0 = 0
            cant1 = 0
            cant2 = 0
            cant14_0 = 0
            cant14_1 = 0
            cant14_2 = 0
            cantL0 = [0,0,0]
            cantL1 = [0,0,0]
            cantL2 = [0,0,0]
            self.total.append(G.number_of_nodes())
            print("Total nodos", G.number_of_nodes())
            for i in G:
                if(G.nodes[i]["Estado"] == 0): #esta sano
                    cant0+=1
                elif(G.nodes[i]["Estado"] == 1): #esta con
                    cant1+=1
                    #cant14_1+=1
                    if(G.nodes[i]["tinfeccion"]<=5):
                        cantL1[0]+=1
                    elif(G.nodes[i]["tinfeccion"]<=10):
                        cantL1[1]+=1
                    elif(G.nodes[i]["tinfeccion"]<=14):
                        cantL1[2]+=1
                else: #esta sin
                    cant2+=1
                    #cant14_2+=1
                    if(G.nodes[i]["tinfeccion"]<=5):
                        cantL2[0]+=1
                    elif(G.nodes[i]["tinfeccion"]<=10):
                        cantL2[1]+=1
                    elif(G.nodes[i]["tinfeccion"]<=14):
                        cantL2[2]+=1
            print("Estados - Sanos:",cant0,"Con:",cant1,"Sin:",cant2)
            self.historico.append([cant1,cant2])
            #print("0:",cant14_0)
            #print("1:",cant14_1)
            #print("2:",cant14_2)
            print("Rangos tiempo infeccion:")
            print("1-5: [",cantL1[0],"] 6-10: [",cantL1[1],"] 11-14: [",cantL1[2],"]")
            print("1-5: [",cantL2[0],"] 6-10: [",cantL2[1],"] 11-14: [",cantL2[2],"]")
            print()
        
    def printGraficos(self):
        self.printNewInfected()
        self.printCantNodos()
        self.printGrados()
        
    def printGrados(self):
        for i in range(len(CH)):
            self.gradoGrafo(CH[i],i)
    
    def printDiameter(self): #En desuso hasta el momento
        listaC = []
        listaV = []
        listaS = []
        for i in range(0,len(self.hist_diametro),3):
            listaC.append(self.hist_diametro[i])
        for i in range(1,len(self.hist_diametro),3):
            listaV.append(self.hist_diametro[i])
        for i in range(2,len(self.hist_diametro),3):
            listaS.append(self.hist_diametro[i])
        self.graficoXdias(listaC,listaV,listaS,"Diametro x dia")
        
    def printCantNodos(self):
        listaC = []
        listaV = []
        listaS = []
        for i in range(0,len(self.total),3):
            listaC.append(self.total[i])
        for i in range(1,len(self.total),3):
            listaV.append(self.total[i])
        for i in range(2,len(self.total),3):
            listaS.append(self.total[i])
        
        self.graficoXdias(listaC,listaV,listaS,"Num nodos")
        
    def printNewInfected(self):
        lista_nuevos = []
        listaC = []
        listaV = []
        listaS = []
        for i in range(0,len(self.hist_nuevo),3):
            listaC.append(self.hist_nuevo[i])
        for i in range(1,len(self.hist_nuevo),3):
            listaV.append(self.hist_nuevo[i])
        for i in range(2,len(self.hist_nuevo),3):
            listaS.append(self.hist_nuevo[i])    
        
        for i in range(len(listaC)):
            lista_nuevos.append([i+1,listaC[i], listaV[i], listaS[i]])
        panda = pd.DataFrame(lista_nuevos,columns = ['dia' , 'Coquimbo', 'Valparaiso', "Santiago"])
        self.grafico2(panda)
        self.graficoXdias(listaC,listaV,listaS,"New infected")
        
    def graficoXdias(self, C, V, S,text):
        lista_supp = range(1,6,1)
        fig,axs = plt.subplots(3, 1)
        axs[0].set_title("Coquimbo")
        axs[0].set_ylabel(text)
        axs[0].set_xlabel('dias')
        axs[0].plot(lista_supp,C)
        
        axs[1].set_title("Valparaiso")
        axs[1].set_ylabel(text)
        axs[1].set_xlabel('dias')
        axs[1].plot(lista_supp,V)
        
        axs[2].set_title("Santiago")
        axs[2].set_ylabel(text)
        axs[2].set_xlabel('dias')
        axs[2].plot(lista_supp,S)
        
        fig.tight_layout()
        plt.show()
        
    def grafico2(self,panda):
        panda[['Coquimbo', 'Valparaiso', "Santiago"]].plot(kind="bar", figsize=(16, 4))
        
    def agregar(self,new):
        self.hist_nuevo.append(new)
        
    def gradoGrafo(self,G,i):
        degree_sequence = sorted([d for n, d in G.degree()], reverse=True)
        degreeCount = collections.Counter(degree_sequence)
        deg, cnt = zip(*degreeCount.items())

        fig, ax = plt.subplots()
        plt.bar(deg, cnt, width=0.80, color="b")
        
        region=""
        if(i==0):region="Coquimbo"
        elif(i==1): region="Valparaiso"
        else: region="Santiago"
        plt.title("Histograma de grados - "+region)
        plt.ylabel("Num personas")
        plt.xlabel("Grado")
        ax.set_xticks([d + 0.4 for d in deg])
        ax.set_xticklabels(deg)

        plt.axes([0.5, 0.5, 0.5, 0.5])
        plt.axis("off")
        plt.show()
        print("Grado promedio del grafo "+region+": ",np.mean(degree_sequence))
        
    #Metodos para dibujar grafo
    def setPositionGraph(self,CH): #guarda posiciones de nodos inicialmente
        for grafo_region in CH:
            self.posGraph.append(nx.layout.spring_layout(grafo_region))
    
    def resetPositionGraph(self,CH): #reajustar los nuevos nodos que llegaron a la region y se fueron
        indice_pos_region=0
        for grafo_region in CH: #recorro las 3 regiones
            list_nodes_actual=list(grafo_region.nodes())
            list_pkeys=list(self.posGraph[indice_pos_region].keys()) #nodos que estaban antes (keys) y su posicion (value)
            
            #resolviendo problema => en mi grafo actual, puede que se hayan ido nodos, debo quitarlo del antiguo
            for antiguo_ele in list_pkeys:
                if antiguo_ele not in list_nodes_actual: #si este elem antiguo, no esta en el actual se elimina de posiciones
                    self.posGraph[indice_pos_region].pop(antiguo_ele)
                    
            oldnodes_con_viaje=list(self.posGraph[indice_pos_region].keys())
            
            #arreglo la posicion de los nodos antiguos, agregando los nuevos nodos y sus posiciones
            self.posGraph[indice_pos_region] = nx.spring_layout(grafo_region, pos=self.posGraph[indice_pos_region], fixed=oldnodes_con_viaje) 
            indice_pos_region+=1
    

    def drawGraph(self,CH,dia): #guardará varias imagenes de la simulación del grafo por día
        #dibujar la region y guardarla: dia_region_grafo.png
        indice_pos_region=0
        regiones=["C","V","S"]
        for g in CH:
            print("Region:",regiones[indice_pos_region])
            lista_color_estado = []
            lista_color_region = []
            for i in g.nodes():
                lista_color_region.append(g.nodes[i]["Color_region"])
                lista_color_estado.append(g.nodes[i]["Color_estado"])
            plt.figure(figsize=(10, 10)) #num=None,, facecolor='w', edgecolor='k')#dpi=80
            nx.draw(g,self.posGraph[indice_pos_region],with_labels=True,font_weight='bold',linewidths=8,node_color = lista_color_region)
            nx.draw(g,self.posGraph[indice_pos_region],with_labels=True,font_weight='bold',node_color = lista_color_estado)
            plt.savefig(str(dia)+"_"+regiones[indice_pos_region]+"_grafo.png",format="PNG")
            plt.show()
            indice_pos_region+=1
            
    #falta poner colores a estados y otras cosas
```

### Inicio de simulación completa - Nota: Se puede agrandar la cantidad de nodos, pero podría tardar bastante
#### Se usa watts_strogatz_graph (pequeño mundo, ya que se asemeja bastante a las redes de las personas)

```python
S = nx.watts_strogatz_graph(355,6, 0.9)#Santiago  --> 355 (numero de nodo - población), 6 (cantidad de aristas maximo), 0.9 (mayor dinamismo en la red) 
V = nx.watts_strogatz_graph(80,6, 0.9)#Valparaiso              
C = nx.watts_strogatz_graph(35,6, 0.9)#Coquimbo                
CH = [C,V,S]

establecerGrafos(CH)
```
